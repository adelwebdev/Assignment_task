```python
# Conduct EDA
# Start by loading the dataset and conducting exploratory data analysis.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_parquet('client_split.parquet', engine='pyarrow')

# Display basic information
print(data.info())
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Visualize distributions of key features
sns.histplot(data[''])

# Visualize correlations
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, fmt='.2f')

# Preprocessing
# Handle missing values
data.fillna(method='ffill', inplace=True)

# Encode categorical variables if necessary
data = pd.get_dummies(data, drop_first=True)
```

# Nouvelle section

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Define features and target for product purchase prediction
X_campaign = data.drop(columns=['product_purchase'])
y_campaign = data['product_purchase']

# Split the data
X_train_campaign, X_test_campaign, y_train_campaign, y_test_campaign =
train_test_split(X_campaign, y_campaign, test_size=0.2, random_state=42)

# Train model for campaign prediction
model_campaign = RandomForestClassifier()
model_campaign.fit(X_train_campaign, y_train_campaign)

# Predictions and evaluation
predictions_campaign = model_campaign.predict(X_test_campaign)
print(classification_report(y_test_campaign, predictions_campaign))

# Define features and target for churn prediction
X_churn = data.drop(columns=['churn'])
y_churn = data['churn']

# Split the data
X_train_churn, X_test_churn, y_train_churn, y_test_churn =
train_test_split(X_churn, y_churn, test_size=0.2, random_state=42)

# Train model for churn prediction
model_churn = RandomForestClassifier()
model_churn.fit(X_train_churn, y_train_churn)

# Predictions and evaluation
predictions_churn = model_churn.predict(X_test_churn)
print(classification_report(y_test_churn, predictions_churn))
```

```python
# Conduct Formal Analysis
# You can analyze the performance metrics obtained from the classification reports.
# Discuss precision, recall, F1-score, and overall accuracy for each model.
# Visualize confusion matrices to better understand model performance.

# Confusion Matrix for Campaign Prediction
cm_campaign = confusion_matrix(y_test_campaign, predictions_campaign)
sns.heatmap(cm_campaign, annot=True)
plt.title('Confusion Matrix for Campaign Prediction')

# Confusion Matrix for Churn Prediction
cm_churn = confusion_matrix(y_test_churn, predictions_churn)
sns.heatmap(cm_churn, annot=True)
plt.title('Confusion Matrix for Churn Prediction')
```
